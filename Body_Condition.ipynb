```python
corelation = diabetic.corr()
```

```python
sns.heatmap(corelation, xticklabels=corelation.columns, yticklabels=corelation.columns ,annot=True)
```

```python
sns.pairplot(diabetic)
```

```python
sns.relplot(x= 'SkinThickness', y='Insulin', hue='BMI', data=diabetic)
```

```python
sns.distplot(diabetic['Pregnancies'], bins=5)
```

```python
sns.catplot(x='BMI', kind= 'box', data= diabetic)
```
